# **ML-MODEL**

# LOAD DATA

```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data Preparation**

## Splitting x and y [y=f(x)]

```python
y=df['logS']
y
```

```python
x=df.drop('logS', axis=1)
x
```

## Data Split for test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# Model Building

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

### Prediction

```python
y_lr_predict=lr.predict(x_train)
y_lr_predict_test=lr.predict(x_test)
```

```python
y_lr_predict
```

```python
y_lr_predict_test
```

###Evaluation


```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse=mean_squared_error(y_train, y_lr_predict)
lr_train_rs=r2_score(y_train,y_lr_predict)

lr_test_mse=mean_squared_error(y_test, y_lr_predict_test)
lr_test_rs=r2_score(y_test,y_lr_predict_test)
```

```python
print('Train MSE:', lr_train_mse)
print('Train rs:', lr_train_rs)
print('Test MSE:', lr_test_mse)
print('Test rs:', lr_test_rs)
```

```python
lr_res=pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_rs, lr_test_mse,lr_test_rs]).transpose()
lr_res.columns=['method','Train MSE','Train RS','Test MSE','Test RS']
```

```python
lr_res
```

## Random Forest

### Prediction


```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_predict=rf.predict(x_train)
y_rf_predict_test=rf.predict(x_test)
```

### Evaluation


```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse=mean_squared_error(y_train, y_rf_predict)
rf_train_rs=r2_score(y_train,y_lr_predict)

rf_test_mse=mean_squared_error(y_test, y_rf_predict_test)
rf_test_rs=r2_score(y_test,y_rf_predict_test)
```

```python
rf_res=pd.DataFrame(['Random Forest', rf_train_mse, rf_train_rs, rf_test_mse,rf_test_rs]).transpose()
rf_res.columns=['method','Train MSE','Train RS','Test MSE','Test RS']
```

```python
rf_res
```

## Model Comparison

```python
df_models=pd.concat([lr_res,rf_res], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```
